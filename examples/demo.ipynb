# quahog demo — milestones 1-3
A live shell embedded in the notebook; every command's output lands as plain text.

**Try minuting:** click into the console below, type a command (e.g. `uname -a`), press Enter.
It's tracked in `h.minutes`, and the cell's own output (its `text/plain`) updates to show it.
Run the `h.dump_minutes_as_cell()` cell to turn everything typed since the last
dump into a new `%qua` cell (works in JupyterLab and VS Code — the cell appears below the
dumping cell).

```python
import quahog as q
h = q.bash()
h  # live, resizable console — type into it!
```

```python
%qua echo hello from quahog
```

```python
%%qua
cd /tmp
pwd
```

```python
h.minutes  # every command typed into the console: when, command, raw/text slices
```

```python
h.dump_minutes_as_cell()  # typed commands since last dump -> new %qua cell below
```

```python
# fork(): fresh stdio, separate stdout/stderr, session stays free
f = h.fork("echo to-stdout; echo to-stderr 1>&2; sleep 0.3; exit 3")
f.wait(10)
f
```

**Recording:** `h.record(True)` starts an asciicast v2 sidecar next to the notebook (`demo.quahog/<session>-<ts>.cast`); the cell above's output notes it while it's active. Full-screen apps (`vim`, `less`, `htop`) get a screenshot instead of garbled text — `clean_text()` can't flatten 2D cursor-addressed output, so it's excluded from the log.

```python
h.record(True)
h.run("echo this command is being recorded")
h.cast_path
```
