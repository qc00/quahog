# quahog demo — milestones 1 + 2
A live shell embedded in the notebook; every command's output lands as plain text.

**Try minuting:** click into the console below, type a command (e.g. `uname -a`), press Enter,
then run any cell — your command comes back as a new `%qua` cell, and its output is appended
to the transcript under the console.

```python
import quahog as q
h = q.bash()
h  # live, resizable console — type into it!
```

```python
%qua echo hello from quahog
```

```python
%qua echo minuted-retest
```

```python
%%qua
cd /tmp
pwd
```

```python
# fork(): fresh stdio, separate stdout/stderr, session stays free
f = h.fork("echo to-stdout; echo to-stderr 1>&2; sleep 0.3; exit 3")
f.wait(10)
f
```
